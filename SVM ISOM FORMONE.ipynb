```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import svm
from sklearn import metrics
from sklearn.metrics import classification_report, accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
traindf = pd.read_csv('train.csv')
testdf = pd.read_csv('valid.csv')
```

```python
traindf.head()
```

```python
trainpre = traindf[['text','funny','cool','useful','stars']]
testpre = testdf[['text','funny','cool','useful','stars']]
trainhaha = trainpre['text']
testhaha = testpre['text']
trainhehe = trainpre['stars']
testhehe = testpre['stars']
```

```python
vectorizer = TfidfVectorizer()
vectorizer.fit_transform(trainhaha)
print('vocabulary: ', vectorizer.vocabulary_)
print('idfs: ', vectorizer.idf_)
vector = vectorizer.transform(testhaha)
print('vectors: ', vector.toarray())
```

```python
train_vectors = vectorizer.fit_transform(trainhaha)
test_vectors = vectorizer.transform(testhaha)
```

```python
classifier_linear = svm.SVC(kernel='linear')
classifier_linear.fit(train_vectors, trainhehe)
prediction_train = classifier_linear.predict(train_vectors)
prediction_test = classifier_linear.predict(test_vectors)
```

```python
print ('testing dataset accuracy: {}'.format(accuracy_score(testhehe, prediction_test)))
print('\n')
print("Classification report for classifier %s:\n%s\n"
      % (classifier_linear, metrics.classification_report(testhehe, prediction_test)))
print("Confusion matrix:\n%s" % metrics.confusion_matrix(testhehe, prediction_test))
```
